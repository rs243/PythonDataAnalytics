Importing libraries and loading csv file

```python
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
df.head(50)
```

Descriptive Statistics

```python
#inspect the basic structure of data
#number of rows and columns (df.shape)
print('Shape (rows, columns):', df.shape)

#column names (df.columns)
print('\nColumn names:', df.columns)

#data types and non-null counts (df.info())
print('\nData types and non-null counts:', df.info())

#summary statistics for numeric columns
print('\nSummary statistics for numeric columns:', df.describe())

#summary statistics for all columns
print('\nSummary statistics for all columns:', df.describe(include='all'))

#finding the missing values
print('\nMissing values:', df.isnull().sum())
```

Cleaning Data

```python
#cleaning column names
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)
df.columns
df.head(10)
```

```python
#replacing the value 's' and 'na' with 'NaN' for columns

#mean_score
df['mean_score'] = df['mean_score'].replace(['s', 'na'], np.nan)

#number_scoring_65_or_above
df['number_scoring_65_or_above'] = df['number_scoring_65_or_above'].replace(['s', 'na'], np.nan)

#percent_scoring_65_or_above
df['percent_scoring_65_or_above'] = df['percent_scoring_65_or_above'].replace(['s', 'na'], np.nan)

#number_scoring_80_or_above
df['number_scoring_80_or_above'] = df['number_scoring_80_or_above'].replace(['s', 'na'], np.nan)

#percent_scoring_80_or_above
df['percent_scoring_80_or_above'] = df['percent_scoring_80_or_above'].replace(['s', 'na'], np.nan)

#number_scoring_below_65
df['number_scoring_below_65'] = df['number_scoring_below_65'].replace(['s', 'na'], np.nan)

#percent_scoring_below_65
df['percent_scoring_below_65'] = df['percent_scoring_below_65'].replace(['s', 'na'], np.nan)

#number_scoring_cr
df['number_scoring_cr'] = df['number_scoring_cr'].replace(['s', 'na'], np.nan)

#percent_scoring_cr
df['percent_scoring_cr'] = df['percent_scoring_cr'].replace(['s', 'na'], np.nan)

df.info()
```

```python
#finding the missing values
print('Missing values:', df.isnull().sum())
```

```python
#filling in the missing values with unknown in regents_exam column to prevent errors
df['regents_exam'] = df['regents_exam'].fillna('unknown')
print('Missing values:', df.isnull().sum())
```

```python
#converting all the object data type which should be numeric into numeric data type
df['mean_score'] = pd.to_numeric(df['mean_score'], errors='coerce')

#number_scoring_65_or_above
df['number_scoring_65_or_above'] = pd.to_numeric(df['number_scoring_65_or_above'], errors='coerce')

#percent_scoring_65_or_above
df['percent_scoring_65_or_above'] = pd.to_numeric(df['percent_scoring_65_or_above'], errors='coerce')

#number_scoring_80_or_above
df['number_scoring_80_or_above'] = pd.to_numeric(df['number_scoring_80_or_above'], errors='coerce')

#percent_scoring_80_or_above
df['percent_scoring_80_or_above'] = pd.to_numeric(df['percent_scoring_80_or_above'], errors='coerce')

#number_scoring_below_65
df['number_scoring_below_65'] = pd.to_numeric(df['percent_scoring_below_65'], errors='coerce')

#percent_scoring_below_65
df['percent_scoring_below_65'] = pd.to_numeric(df['percent_scoring_below_65'], errors='coerce')

#number_scoring_cr
df['number_scoring_cr'] = pd.to_numeric(df['number_scoring_cr'], errors='coerce')

#percent_scoring_cr
df['percent_scoring_cr'] = pd.to_numeric(df['percent_scoring_cr'], errors='coerce')

df.info()
```

```python
#converting columns into category
categorical_cols = ['school_name', 'school_level', 'regents_exam', 'year']
df[categorical_cols] = df[categorical_cols].astype('category')
df[categorical_cols].dtypes
```

Descriptive Statistics

```python
#descriptive statistics for numeric columns and categorical columns
print("\nDescriptive statistics for numeric columns:")
print(df.describe())

print("\nValue counts for school_name:")
print(df['school_name'].value_counts())

print("\nValue counts for school_level:")
print(df['school_level'].value_counts())

print("\nValue counts for regents_exam:")
print(df['regents_exam'].value_counts())

print("\nValue counts for year:")
print(df['year'].value_counts())
```

1. Pick any school in these dataset as initial comparison point: Long Island City High School and Park East High School
2. Pick a feature/measure/score: Mean Score

```python
#bivariate analysis for long island city high school and harry truman high school
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['school_name'] == 'Long Island City High School']['mean_score'], label='Long Island City', fill=True, color='green')
sns.kdeplot(df[df['school_name'] == 'Park East High School']['mean_score'], label='Park East High School', fill=True, color='red')
plt.title('Combined Distribution of Mean Score: Long Island City High School vs Park East High School')
plt.xlabel('Mean Score')
plt.ylabel('Density')
plt.legend()
plt.show()



```

In the plot we can see differences in the mean score distributions for these two schools. 'Long Island City High School' have higher peak mean scores compared to 'Park East High School' which have wider spread of mean scores. The plot suggests that 'Long Island City High School' have stronger performance than 'Park East High School.
